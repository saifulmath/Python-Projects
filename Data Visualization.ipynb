                                       Data Visualization with Python

In this section, I have represented discrete data visualization.Here, I choose Cars data,Real estate company data and KD Nuggets Survey data to visualize Bar chart,Pie chart, Stacked Area chart,Line chart,Histogram, Scatter plot,Regession plot,Combine Bar & Line chart.

```python
# Bar Chart
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set() # to set background shade of figure
```

```python
df_used_cars = pd.read_csv("bar_chart_data.csv")
```

```python
df_used_cars
```

```python
plt.figure(figsize = (9,6))
plt.bar(x = df_used_cars["Brand"],height = df_used_cars["Cars Listings"])
plt.show()
```

```python
plt.bar(x = df_used_cars["Brand"],height = df_used_cars["Cars Listings"])
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (9,6))
colors = ['red', 'green', 'blue', 'white', 'yellow', 'magenta', 'cyan']
plt.bar(x = df_used_cars["Brand"],height = df_used_cars["Cars Listings"],color = colors)
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (9,6))
#colors = ['red', 'green', 'blue', 'white', 'yellow', 'magenta', 'cyan']
plt.bar(x = df_used_cars["Brand"],height = df_used_cars["Cars Listings"],color = "midnightblue")
plt.xticks(rotation = 45)
plt.title("Cars Listings by Brand")
plt.show()
```

```python
plt.figure(figsize = (9,6))
plt.bar(x = df_used_cars["Brand"],height = df_used_cars["Cars Listings"],color = "midnightblue")
plt.xticks(rotation = 45,fontsize = 13)
plt.yticks(fontsize = 13)
plt.title("Cars Listings by Brand",fontsize = 16,fontweight = "bold")
plt.ylabel("Number of Listings", fontsize = 13)
plt.show()
plt.savefig("Used Cars Bar.png")
```

```python
# Pie Chart
```

```python
df_fuel_engine_type = pd.read_csv("pie_chart_data.csv")
```

```python
df_fuel_engine_type
```

```python
plt.figure(figsize = (10,8))
plt.pie(x = df_fuel_engine_type['Number of Cars'],
        labels = df_fuel_engine_type['Engine Fuel Type'].values,autopct = '%.2f')
plt.show()
```

```python
sns.set_palette('colorblind')
```

```python
plt.figure(figsize = (10,8))
plt.pie(x = df_fuel_engine_type['Number of Cars'],
        labels = df_fuel_engine_type['Engine Fuel Type'].values,autopct = '%.2f',
       textprops = {'size': 'x-large','fontweight': 'bold',
                    'rotation':30,'color':'white'}) # 'textprops' is a dictionary of arguments for text objects
plt.legend()
plt.title('Cars by Engine Fuel Type',fontsize = 18, fontweight = 'bold')
plt.show()
```

```python
# Stacked Area Chart
```

```python
df_fuel_engine_types = pd.read_csv("stacked_area_chart_data.csv")
```

```python
df_fuel_engine_types
```

```python
plt.figure(figsize = (12,6))
plt.stackplot(df_fuel_engine_types["Year"],df_fuel_engine_types["Diesel"],df_fuel_engine_types["Petrol"],
              df_fuel_engine_types["Gas"])
plt.xticks(df_fuel_engine_types["Year"],rotation = 45)
plt.show()                                  
```

```python
colors = ["#011638","#7e2987","#ef2026"] # color code for dark blue,violet,pinkish red
plt.figure(figsize = (12,6))
plt.stackplot(df_fuel_engine_types["Year"],df_fuel_engine_types["Diesel"],df_fuel_engine_types["Petrol"],
              df_fuel_engine_types["Gas"],colors = colors)
plt.xticks(df_fuel_engine_types["Year"],rotation = 45)
plt.show() 
```

```python
colors = ["#011638","#7e2987","#ef2026"] # color code for dark blue,violet,pinkish red
labels = ["Diesel","Petrol","Gas"]
sns.set_style("white") # set for background
plt.figure(figsize = (12,6))
plt.stackplot(df_fuel_engine_types["Year"],df_fuel_engine_types["Diesel"],
              df_fuel_engine_types["Petrol"],df_fuel_engine_types["Gas"],
              colors = colors,edgecolor = 'None') # here,remove white edgecolor middle of  each graph
plt.xticks(df_fuel_engine_types["Year"],rotation = 45)
plt.legend(labels = labels,loc = "upper left")
plt.ylabel("Number of Cars", fontsize = 13)
plt.title("Popularity of Engine Fuel Types (1982 - 2016)",fontsize = 14,weight = "bold")
sns.despine() #write sns dspine to remove the top and right borders of our chart.We can specifically remove a grid with left, right, bottom,or top and equal to true.
plt.show() 
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set() # to set background shade of figure
```

```python
df_spx_ftse_00_10
```

```python
df_spx_ftse_00_10["new_date"]
```

```python
plt.plot(df_spx_ftse_00_10["new_date"],df_spx_ftse_00_10["GSPC500"])
plt.show()
```

```python
plt.plot(df_spx_ftse_00_10["new_date"],df_spx_ftse_00_10["GSPC500"])
plt.plot(df_spx_ftse_00_10["new_date"],df_spx_ftse_00_10["FTSE100"])
plt.show()
```

```python
# For better viz,choose small timeframe & slice the data.
df_spx_ftse_H2_08 = df_spx_ftse_00_10[(df_spx_ftse_00_10.new_date >= '2008-07-01') & 
                                     (df_spx_ftse_00_10.new_date <= '2008-12-31')]
```

```python
df_spx_ftse_H2_08
```

```python
#labels = ["S&P 500","FTSE 100"]
plt.figure(figsize = (20,8))
plt.plot(df_spx_ftse_H2_08["new_date"],df_spx_ftse_H2_08["GSPC500"], color = "midnightblue")
plt.plot(df_spx_ftse_H2_08["new_date"],df_spx_ftse_H2_08["FTSE100"],color = "crimson")
plt.title("S&P vs FTSE Returns (H2 2008)",fontsize = 14, fontweight = "bold")
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels,fontsize = "large")
plt.show()
```

```python
# Scatter Plot
```

```python
df_real_estate = pd.read_csv("scatter_data.csv")
```

```python
df_real_estate
```

```python
plt.figure(figsize = (12,8))
plt.scatter(df_real_estate['Area (ft.)'],df_real_estate['Price'])
plt.show()
```

```python
plt.figure(figsize = (12,8))
plt.scatter(df_real_estate['Area (ft.)'],
            df_real_estate['Price'],alpha = 0.6) # alpha represents transparency,value (0-1)
plt.show()
```

```python
plt.figure(figsize = (12,8))
plt.scatter(df_real_estate['Area (ft.)'],df_real_estate['Price'],
            alpha = 0.6,c = df_real_estate['Building Type'],
           cmap = 'viridis')# cmap means color map
plt.show()
```

```python
plt.figure(figsize = (12,8))
scatter = plt.scatter(df_real_estate['Area (ft.)'],
                      df_real_estate['Price'],
                      alpha = 0.6,c = df_real_estate['Building Type'],
                      cmap = 'viridis')# cmap means color map
plt.legend(*scatter.legend_elements(),loc = "upper left",title = "Building Type")
plt.title("Realationship in Area and Price",fontsize = 14,weight = "bold")
plt.xlabel("Area (sq. ft.)", weight = "bold")
plt.ylabel("Price (000's of $)")
plt.show()
```

```python
# Scatter plot using Seaborn 
plt.figure(figsize = (12,8))
sns.scatterplot(x = df_real_estate['Area (ft.)'],
                y = df_real_estate['Price'],
                hue = df_real_estate["Building Type"],
                palette = ['black','darkblue','purple','pink','green'],
                s = 100)#s means size of dot
plt.title("Realationship in Area and Price",fontsize = 14,weight = "bold")
plt.xlabel("Area (sq. ft.)", weight = "bold")
plt.ylabel("Price (000's of $)")
plt.show()
```

```python
# Regression Plot
```

```python
df_ad_expenditure = pd.read_csv("scatter_plot_ii(Regression).csv")
```

```python
df_ad_expenditure
```

```python
plt.figure(figsize = (10,8))
sns.regplot(x = "Budget",
            y = "Sales",
            data = df_ad_expenditure)
plt.show()
```

```python
#plt.figure(figsize = (10,8))
sns.set(rc = {'figure.figsize': (9,6)})
sns.regplot(x = "Budget",
            y = "Sales",
            data = df_ad_expenditure,     # kws means Keyword arguments.
            scatter_kws = {'color': 'k'}, #Sets the color of the scatter plot points to black
            line_kws = {'color':'red'})   # Sets the color of the regression line to red
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's $)")
plt.title("Effect of Ad Expenditure on Salses", fontsize = 14, weight = "bold")
plt.show()
```

```python
sns.lmplot(x = "Budget",
            y = "Sales",
            data = df_ad_expenditure ,height = 10, # kws means Keyword arguments.
            scatter_kws = {'color': 'k'}, #Sets the color of the scatter plot points to black
            line_kws = {'color':'red'})   # Sets the color of the regression line to red
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's $)")
plt.title("Effect of Ad Expenditure on Salses", fontsize = 14, weight = "bold")
plt.show()
```

```python
import  pandas as pd 
import matplotlib.pyplot as plt 

from matplotlib.ticker import PercentFormatter 
```

```python
df_kdnuggets = pd.read_csv("bar_line_chart_data.csv")
```

```python
df_kdnuggets
```

```python
fig, ax = plt.subplots(figsize = (10,7))

ax.bar(df_kdnuggets["Year"],
       df_kdnuggets["Participants"],color = "k")
```

```python
fig, ax = plt.subplots(figsize = (10,7))

ax.bar(df_kdnuggets["Year"],
       df_kdnuggets["Participants"],color = "k")
ax1 = ax.twinx()
ax1.plot(df_kdnuggets["Year"],
       df_kdnuggets["Python Users"],color = "darkred", marker = 'D') # here, D means Diamond shape
```

```python
fig, ax = plt.subplots(figsize = (10,7))
ax.bar(df_kdnuggets["Year"],
       df_kdnuggets["Participants"],color = "lightgreen",label = "Participants")
ax.set_ylabel("Number of Participants",weight = "bold",color = "lightgreen")
ax.tick_params(axis = "y",width = 2,labelsize = 'large',colors = "lightgreen")
ax1 = ax.twinx()
ax1.set_ylim(0,1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0))
ax1.plot(df_kdnuggets["Year"],
         df_kdnuggets["Python Users"],color = "darkred", marker = 'D',label = 'Python Users')
ax1.set_ylabel("Python Users",weight = "bold",color = 'red')
ax1.tick_params(axis = "y",width = 2,labelsize = 'large',colors = 'red')
ax.set_title("KD Nuggets Survey Python Users (2012 - 2019)",fontsize = "14",weight = "bold")
ax.legend(loc = 'upper left',fontsize = "large")
ax1.legend(loc = 'upper right' ,fontsize = "large")
```
